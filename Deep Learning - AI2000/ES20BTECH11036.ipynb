```python
def generate_data(no_points):
    X = np.zeros(shape = (no_points,2))
    Y = np.zeros(shape = no_points)
    for i in range(no_points):
        X[i][0] = np.random.randint(1,100)+0.5
        X[i][1] = np.random.randint(1,100)+0.5
        Y[i] = 1 if 7* X[i][0] -5* X[i][1] > 13 else -1
    return X, Y
```

```python
x, y = generate_data(1000)
```

```python
for i in range(len(x)):
    if y[i] == 1:
        plt.scatter(x[i,0], x[i,1], c = 'r') 
    elif y[i] == -1:
        plt.scatter(x[i,0], x[i,1], c = 'g')
    else:
        plt.scatter(x[i,0], x[i,1], c = 'b')
        
plt.show()
```

```python
p = Perceptron()
p.fit(x,y)

x_test, ytest = generate_data(50)
predicted_y_test = p.classify(x_test)
```
